```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation
%matplotlib notebook
```

```python
plt.style.use('fivethirtyeight')
```

```python
frame_len = 10000

fig = plt.figure(figsize=(9,6))

def animate(i):
    data = pd.read_csv('tweets_sentiment.csv')
    y1 = data['investec']
    y2 = data['absa']
    y3 = data['nedbank']
    y4 = data['discovery_bank']
    y5 = data['capitec']
    y6 = data['africa_bank']
    y7 = data['tyme_bank']
    y8 = data['bank_zero']
    y9 = data['standard_bank']
    

    if len(y1)<=frame_len:
        plt.cla()
        plt.plot(y1, label='investec')
        plt.plot(y2, label='absa')
        plt.plot(y3, label='nedbank')
        plt.plot(y4, label='discovery_bank')
        plt.plot(y5, label='capitec')
        plt.plot(y6, label='africa_bank')
        plt.plot(y7, label='tyme_bank')
        plt.plot(y8, label='bank_zero')
        plt.plot(y9, label='standard_bank')
    else:
        plt.cla()
        plt.plot(y1[-frame_len: ], label='investec')
        plt.plot(y2[-frame_len: ], label='absa')
        plt.plot(y3[-frame_len: ], label='nedbank')
        plt.plot(y4[-frame_len: ], label='discovery_bank')
        plt.plot(y4[-frame_len: ], label='capitec')
        plt.plot(y4[-frame_len: ], label='africa_bank')
        plt.plot(y4[-frame_len: ], label='tyme_bank')
        plt.plot(y4[-frame_len: ], label='bank_zero')
        plt.plot(y4[-frame_len: ], label='standard_bank')
    
    plt.legend(loc='upper left')
    plt.tight_layout()


ani = FuncAnimation(plt.gcf(), animate, interval=1000)
```
