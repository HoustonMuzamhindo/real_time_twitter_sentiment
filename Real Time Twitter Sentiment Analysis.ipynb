```python
import tweepy
from tweepy import Stream
from tweepy import StreamListener
import json
from textblob import TextBlob
```

```python
import re
import csv
import contractions
```

```python
investec = 0
absa = 0
nedbank = 0
discovery_bank = 0
capitec = 0
africa_bank = 0
tyme_bank = 0
bank_zero = 0
standard_bank = 0

header_name = ['investec', 'absa', 'nedbank', 'discovery_bank', 'capitec', 'africa_bank', 'tyme_bank', 'bank_zero', 'standard_bank']
with open('tweets_sentiment.csv', 'w') as file:
    writer = csv.DictWriter(file, fieldnames=header_name)
    writer.writeheader()

class Listener(StreamListener):
    
    def on_data(self, data):
        raw_twitts = json.loads(data)
        try:
            tweets = raw_twitts['text']
            
            #tweets = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweets).split())
            tweets = ' '.join(re.sub('RT',' ', tweets).split())
            
            #clean the sentence a bit... remove contractions and also put it to lower and remove white space then create a blob
            blob = TextBlob(contractions.fix(tweets.strip().lower()))
            
            global investec
            global absa
            global nedbank
            global discovery_bank
            global capitec
            global africa_bank
            global tyme_bank
            global bank_zero
            global standard_bank
            
            investec_sentiment = 0
            absa_sentiment = 0
            nedbank_sentiment = 0
            discovery_bank_sentiment = 0
            capitec_sentiment = 0
            africa_bank_sentiment = 0
            tyme_bank_sentiment = 0
            standard_bank_sentiment = 0
            bank_zero_sentiment = 0
            
            for sent in blob.sentences:
                
                #print the sentence and the polarity
                print(sent,"....", sent.sentiment.polarity)
                print()
                
                if "investec" in sent:
                    investec_sentiment = investec_sentiment + sent.sentiment.polarity
                elif "absa" in sent:
                    absa_sentiment = absa_sentiment + sent.sentiment.polarity
                elif "nedbank" in sent:
                    nedbank_sentiment = nedbank_sentiment + sent.sentiment.polarity
                elif "discovery" in sent:
                    discovery_sentiment = discovery_sentiment + sent.sentiment.polarity
                elif "capitec" in sent:
                    capitec_sentiment = capitec_sentiment + sent.sentiment.polarity
                elif "africa bank" in sent:
                    africa_bank_sentiment = africa_bank_sentiment + sent.sentiment.polarity
                elif "tymebank" in sent or "tyme bank" in sent:
                    tyme_bank_sentiment = tyme_bank_sentiment + sent.sentiment.polarity
                elif "standard bank":
                    standard_bank_sentiment = standard_bank_sentiment + sent.sentiment.polarity
                else:
                    bank_zero_sentiment = bank_zero_sentiment + sent.sentiment.polarity
            
            investec = investec + investec_sentiment
            absa = absa + absa_sentiment
            nedbank = nedbank + nedbank_sentiment
            discovery_bank = discovery_bank + discovery_bank_sentiment
            capitec = capitec + capitec_sentiment
            africa_bank = africa_bank + africa_bank_sentiment
            tyme_bank = tyme_bank + tyme_bank_sentiment
            standard_bank_bank = standard_bank + standard_bank_sentiment
            bank_zero = bank_zero + bank_zero_sentiment
            
            with open('tweets_sentiment.csv', 'a') as file:
                writer = csv.DictWriter(file, fieldnames=header_name)
                info = {
                    'investec': investec,
                    'absa': absa,
                    'nedbank' : nedbank,
                    'discovery_bank' : discovery_bank,
                    'capitec' : capitec,
                    'africa_bank' : africa_bank,
                    'tyme_bank' : tyme_bank,
                    'bank_zero' : bank_zero,
                    'standard_bank' : standard_bank
                }
                writer.writerow(info)
            
            #print(raw_twitts['text'])
            #print()
        except:
            print('Found an Error')
        
    def on_error(self, status):
        print(status)
```

```python
#authenticate
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
```

```python
#create a twitter stream
twitter_stream = Stream(auth, Listener())

#the terms we want to track for the tweets
twitter_stream.filter(track = ['Investec', 'ABSA', 'Nedbank', 'Discovery Bank', 'Capitec', 'Africa Bank', 'Tyme Bank', 
                               'TymeBank', 'Bank Zero', 'BankZero', 'Standard Bank'])
```
